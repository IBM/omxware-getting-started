### Explore genes of interest

```python
search_string = 'sporulation'
response = omx.genes(gene_name=search_string, page_size=25)
```

```python
response.show_facets(name='genera', topN=7)
```

```python
genome_ids_list = ['ERR1216979', 'ERR1212028', 'SRR3314498']
genomes = omx.genomes(ids=genome_ids_list).results()

for genome in genomes:
    omxid = genome.id()
    typ = genome.genome_type()
    taxid = genome.taxid()
    metadata = genome.metadata('biosample')

    genera = genome.genus().total_results()

    print('ID: ' + omxid)
    print('Type: ' + typ)
    print('TaxID: ' + taxid)
    print('BioSample: ' + str(metadata)[:100])
    print('Genera: ' + str(genera) + ' Genus')
    print('Json:' + str(genome.json())[:100])

    #output_file_path = '/you/preferred/output/path/' + genome.id() + '.txt'
    #genome.file(output_file_path=output_file_path)
    
    print("\n")
    
```
